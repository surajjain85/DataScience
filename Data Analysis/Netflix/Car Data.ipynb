# find a null value in a dataset . if there is any null value in any column then fill it with mean of that column

```python
car.head()
```

```python
car.shape
```

```python
car.isnull().sum()
```

```python
#so we have null values in each colum so better to iterate the columns with mean values

#car.columns #to check the column names
na_data_mean= ['EngineSize', 'Cylinders', 'Horsepower', 'MPG_City', 'MPG_Highway',
       'Weight', 'Wheelbase', 'Length']
na_data_mode=['Make', 'Model', 'Type', 'Origin', 'DriveTrain', 'MSRP', 'Invoice']
```

```python
car.dtypes
```

```python
car['Make'].fillna(car['Make'].mode())
```

```python
for i in na_data_mode:
    car[i].fillna(car[i].mode()[0], inplace=True)
    #loop is create for object value with mode at 0th position
```

```python
#now check if the mode is placed on the null values
car.isnull().sum()
```

```python
#now we'll do the same for float values and placing null withmean values
for i in na_data_mean:
    car[i].fillna(car[i].mean(),inplace=True)
```

```python
#now check if the mode is placed on the null values
car.isnull().sum()
```
